```python
from timeit import default_timer as timer
import geopandas as gpd
import pandas as pd
import requests
import json
from shapely.geometry import Polygon, Point
import numpy as np
from osmxtract import overpass, location
from glob import glob

from geopy.geocoders import Nominatim
geolocator = Nominatim(user_agent='samuelfraiberger@gmail.com')
# 

from geopy.extra.rate_limiter import RateLimiter
from functools import partial
from tqdm import tqdm
tqdm.pandas()

%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
```

# Mexico City mxdf_cuadrantes

```python
# https://blog.diegovalle.net/2016/10/crime-maps-of-mexico-city.html
mxdf_cuadrantes = pd.DataFrame(requests.get("https://hoyodecrimen.com/api/v1/mxdf_cuadrantes/geojson").json())
mxdf_cuadrantes = pd.DataFrame(list(mxdf_cuadrantes['features'])).drop('type',1)
mxdf_cuadrantes = pd.concat([
pd.DataFrame(list(mxdf_cuadrantes['properties'])),
pd.DataFrame(list(mxdf_cuadrantes['geometry']))['coordinates']],1)
mxdf_cuadrantes['coordinates'] = mxdf_cuadrantes['coordinates'].apply(lambda x:x[0])
mxdf_cuadrantes['coordinates'] = mxdf_cuadrantes['coordinates'].apply(lambda x:[y[:2] for y in x])
mxdf_cuadrantes['coordinates'] = mxdf_cuadrantes['coordinates'].apply(lambda x:Polygon(np.array(x)))
mxdf_cuadrantes.rename(columns={'coordinates':'polygon'},inplace=True)
mxdf_cuadrantes = gpd.GeoDataFrame(mxdf_cuadrantes,geometry='polygon')
print('# mxdf_cuadrantes', mxdf_cuadrantes.shape[0])
```

```python
# Not Sure How Robust This Is. Perhaps Use a Better Definition Moving Forward.
minx = mxdf_cuadrantes.geometry.bounds.min()['minx']
maxx = mxdf_cuadrantes.geometry.bounds.max()['maxx']
miny = mxdf_cuadrantes.geometry.bounds.min()['miny']
maxy = mxdf_cuadrantes.geometry.bounds.max()['maxy']
```

```python
mxdf_cuadrantes.to_pickle('../data/mxdf_cuadrantes.pkl')
mxdf_cuadrantes.to_file('../data/mxdf_cuadrantes.shp', driver='ESRI Shapefile')
```

# POIs From Twitter

```python
print('Import:')
start = timer()

files = glob('../data/decahose/parsed/locations/places.parquet/*.snappy.parquet')

poi_twitter = pd.DataFrame()
for file in files:
    poi_twitter = pd.concat([poi_twitter,pd.read_parquet(file)])

end = timer()
print('Computing Time:', round(end - start), 'sec')
```

```python
print('# Twitter Places',poi_twitter.shape[0])

# Select POIs in Mexico
poi_twitter = poi_twitter[(poi_twitter['place_type']=='poi')].copy()
print('# Twitter Pois',poi_twitter.shape[0])

poi_twitter = poi_twitter[(poi_twitter['country_code']=='MX')].copy()
print('# Twitter Pois in Mexico',poi_twitter.shape[0])

poi_twitter.dropna(inplace=True)

# Geometry
poi_twitter['coordinates'] = \
poi_twitter['coordinates'].apply(lambda x:Polygon([list(y) for y in x[0]]))
poi_twitter = gpd.GeoDataFrame(poi_twitter,geometry='coordinates')

# Actually Only One Point So Collapse
if not poi_twitter.coordinates.area.sum():
    poi_twitter['coordinates'] = poi_twitter['coordinates'].centroid

poi_twitter.drop(['url','place_type','type','country','country_code'],1,errors='ignore',inplace=True)

# Merge
poi_twitter = gpd.sjoin(poi_twitter,mxdf_cuadrantes[['polygon']],how='inner').drop('index_right',1)
print('# Twitter Pois in Mexico City',poi_twitter.shape[0])

poi_twitter.sort_values(by='n_obs',ascending=False,inplace=True)
poi_twitter.reset_index(drop=True,inplace=True)
```

```python
poi_twitter.to_pickle('../data/poi_twitter.pkl')
poi_twitter.to_file('../data/poi_twitter.shp', driver='ESRI Shapefile')
```

# Geocode Twitter POIs

```python
geocode = RateLimiter(geolocator.geocode, min_delay_seconds=1)
poi_twitter['geocode_osm'] = \
poi_twitter['full_name'].progress_apply(partial(geocode, addressdetails=True, viewbox=(minx,miny,maxx,maxy)))
```

# POIs From OSM

```python
# https://github.com/yannforget/osmxtract
# lat, lon = location.geocode('Mexico City, Mexico')
# bounds = location.from_buffer(lat, lon, buffer_size=20000) # buffer_size in Meters
bounds = (miny,minx,maxy,maxx)

# Build an overpass QL query and get the JSON response
query = overpass.ql_query(bounds, tag='amenity')
response = overpass.request(query)
```

```python
poi_osm = pd.DataFrame(response['elements'])
print('# POI OSM:', poi_osm.shape[0])

poi_osm['amenity_type'] = poi_osm['tags'].apply(lambda x:x.get('amenity',None))
poi_osm['amenity_name'] = poi_osm['tags'].apply(lambda x:x.get('name',None))

# Nodes list elements contained in ways
# Members list elements contained in relations
# Droping Geometry. Using Bounding Box Instead.
poi_osm.drop(['tags','nodes','members','geometry'],1,inplace=True,errors='ignore')

poi_osm['lat'] = poi_osm.apply(
lambda x:(x['bounds']['minlat']+x['bounds']['maxlat'])/2 if type(x['bounds'])==dict else x['lat'],1)

poi_osm['lon'] = poi_osm.apply(
lambda x:(x['bounds']['minlon']+x['bounds']['maxlon'])/2 if type(x['bounds'])==dict else x['lon'],1)

poi_osm['point'] = poi_osm.apply(lambda x:Point(x['lon'],x['lat']),1)

poi_osm = gpd.GeoDataFrame(poi_osm,geometry='point')

poi_osm = gpd.sjoin(poi_osm,mxdf_cuadrantes[['polygon']],how='inner').drop('index_right',1)
poi_osm = poi_osm.sort_values(by='amenity_type').reset_index(drop=True)
print('# POI OSM in Mexico City:', poi_osm.shape[0])
```

```python
poi_osm.to_pickle('../data/poi_osm.pkl')
poi_osm.to_file('../data/poi_osm.shp', driver='ESRI Shapefile')
```

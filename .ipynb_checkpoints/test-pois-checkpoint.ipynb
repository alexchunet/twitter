```python
import folium 
import webbrowser
import pandas as pd
import numpy as np
from glob import glob
from timeit import default_timer as timer
from shapely.geometry import Polygon, Point
from folium.plugins import FastMarkerCluster
import seaborn as sns
import matplotlib.pyplot as plt
import geopandas as gpd
```

```python
print('Import:')
start = timer()

files = glob('../data/decahose/parsed/locations/places.parquet/*.snappy.parquet')

pois = pd.DataFrame()

for file in files:
    
    pois = pd.concat([pois,pd.read_parquet(file)])

pois = pois[(pois['place_type']=='poi')].copy()
pois.drop(['id','url','place_type','type','name'],1,errors='ignore',inplace=True)
pois.dropna(inplace=True)

pois['coordinates'] = pois['coordinates'].apply(lambda x:Polygon([list(y) for y in x[0]]))
pois['coordinates'] = pois['coordinates'].apply(lambda x:x.centroid)
pois['longitude'] = pois['coordinates'].apply(lambda x:x.x)
pois['latitude'] = pois['coordinates'].apply(lambda x:x.y)

pois = pois.sort_values(by='n_obs',ascending=False).reset_index(drop=True)

end = timer()
print('Computing Time:', round(end - start), 'sec')
```

```python
pois.head()
```

```python
# Geodataframe of POIS
geometry = [Point(xy) for xy in zip(pois.longitude, pois.latitude)]
crs = {'init': 'epsg:4326'}
pois_gdf = gpd.GeoDataFrame(pois, crs=crs, geometry=geometry)
```

```python
world = gpd.read_file('/anaconda3/pkgs/geopandas-0.4.1-py_0/site-packages/geopandas/datasets/naturalearth_lowres/naturalearth_lowres.shp')
```

```python
pois_world = gpd.sjoin(pois_gdf, world)
```

```python
def barplot_pois_by_country():

    with sns.axes_style("white"):
        sns.set_style("ticks")
        sns.set_context("talk")

        fig,ax = plt.subplots(nrows=1,ncols=1,figsize=(40,10))

        to_plot = pois_world[pois_world['iso_a3']!='-99']['name'].value_counts()
        to_plot.index = to_plot.index.map(lambda x:x.title().replace('_',' '))

        to_plot.plot(
        ax=ax,
        kind='bar',
        color=sns.color_palette("Set2",10))

        ax.tick_params(which='both',direction='in',pad=3)
        ax.locator_params(axis='y',nbins=6)
        ax.set_xticklabels(ax.get_xticklabels(),rotation=45,ha='right')
        ax.set_yscale('log')

    plt.savefig('../fig/barplot-pois-by-country.pdf',bbox_inches='tight')
    
barplot_pois_by_country()
```

```python
def barplot_pois_by_continent():

    with sns.axes_style("white"):
        sns.set_style("ticks")
        sns.set_context("talk")

        fig,ax = plt.subplots(nrows=1,ncols=1,figsize=(30,10))

        to_plot = pois_world[pois_world['iso_a3']!='-99']['continent'].value_counts()
        to_plot.index = to_plot.index.map(lambda x:x.title().replace('_',' '))

        to_plot.plot(
        ax=ax,
        kind='bar',
        color=sns.color_palette("Set2",10))

        ax.tick_params(which='both',direction='in',pad=3)
        ax.locator_params(axis='y',nbins=6)
        ax.set_xticklabels(ax.get_xticklabels(),rotation=45,ha='right')
        ax.set_yscale('log')

    plt.savefig('../fig/barplot-pois-by-continent.pdf',bbox_inches='tight')
    
barplot_pois_by_continent()
```

# Test Specific Cities

```python
from glob import glob
sorted(glob('../data/shp/*/*AOI.shp'))[0].split('/')[-2]
```

```python
def get_pois_by_city(path):
    
    shp_city = gpd.read_file(path)
    shp_city = shp_city.to_crs({'init': 'epsg:4326'})
    pois_city= gpd.sjoin(pois_gdf, shp_city)
    pois_city=pois_city[['full_name','n_obs','latitude','longitude']].copy()
    pois_city.rename(columns={'full_name':'name','n_obs':'n_checkins'},inplace=True)
    pois_city = pois_city.sort_values(by=['n_checkins','name'],ascending=[False,True]).reset_index(drop=True)
    
    return pois_city


pois_by_city = {}

paths = sorted(glob('../data/shp/*/*AOI.shp'))

for path in paths:
    
    print('Import:')
    start = timer()
    print(path.split('/')[-2])
    pois_by_city[path.split('/')[-2]] = get_pois_by_city(path)

    end = timer()
    print('Computing Time:', round(end - start), 'sec')
```

```python
for city in pois_by_city:
    pois_by_city[city].head(50).to_csv('../data/pois/'+city.split('_')[0]+'.csv')
```

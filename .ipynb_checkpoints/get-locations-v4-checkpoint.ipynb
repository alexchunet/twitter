```python
import sys
import os
import pandas as pd
import geograpy
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
version_locations = 'v4'
version_users     = 'v4'

path_to_locations = './data/decahose/parsed/locations/'
path_to_users     = './data/decahose/parsed/users/'

countries = [
'Argentina',
'Brazil',
'Chile',
'Colombia',
'Indonesia',
'Malaysia',
'Myanmar',
'Pakistan',
'Paraguay',
'Peru',
'Philippines',
'Thailand',
]
```

```python
%%time
print('Import users locations...\n')
users_locations = pd.read_pickle(path_to_users+'users-'+version_users+'.pkl')
print('# Geolocated Users:', users_locations.shape[0])
```

```python
users_locations.head()
```

```python
%%time
freq_locations = users_locations['USER LOCATION'].value_counts().drop('',errors='ignore').rename('N')
```

```python
%%time
def get_candidates(country,cutoff_significance=50,cutoff_likelihood=0.5):
    
    # Select Country Locations
    candidates = users_locations.loc[
    users_locations['COUNTRY']==country,'USER LOCATION'].value_counts().drop('',errors='ignore').rename('N')
    
    # Compute Country Frequencies
    candidates = pd.concat([candidates,
    candidates.div(freq_locations.loc[candidates.index]).rename('RATIO')],1)
    
    # Select Those Above Significance Threshold
    candidates = candidates.loc[candidates['N']>=cutoff_significance].copy()
    
    # Extract All Places
    candidates['PLACES'] = candidates.index.map(lambda x:geograpy.get_place_context(text=x.title()))
    
    candidates['HAS COUNTRY'] = candidates['PLACES'].apply(
    lambda x:int(list(set(x.countries).union(set(x.country_cities.keys()))) == [country]))

    # Only Keep If Appears At Lease 50% of Time in the Country
    candidates = candidates.loc[candidates['RATIO']>cutoff_likelihood].copy()

    return candidates

candidates = {}
print('Extract Locations:')
for country in countries:
    
    candidates[country] = get_candidates(country)
    
    print(country, ': # Candidates =', candidates[country].shape[0])
```

```python
to_drop = [
'na merda',
'Planeta Terra',
'In your timeline',
'WORLD',
'Stamford Bridge, London',
'MANCHESTER UNITED',
'Giuseppe Meazza',
'Jesus Christ',
'Seoul,Korea Selatan',
'Korea selatan',
'Undetected',
'Everywhere~',
'Santiago bernabeu',
'Somewhere down the road',
]

print('Select Locations:')

for country in countries:
    
    candidates[country].drop(to_drop,errors='ignore',inplace=True)
    
    # Keep All Locations Appearing More Than 85% of the time in the Country
    # Otherwise Make sure that the Name
    candidates[country] = candidates[country].loc[
    (candidates[country]['RATIO']>0.85)|(candidates[country]['HAS COUNTRY']==1)].copy()
    
    print(country, ': # Candidates =', candidates[country].shape[0])
```

```python
for country in countries:
    fig,ax = plt.subplots()
    candidates[country].sort_values(by='RATIO')['RATIO'].plot(ax=ax,marker='.',markersize=1)
    candidates[country].sort_values(by='RATIO')['N'].plot(ax=ax,secondary_y=True)
    ax.set_title(country)
```

```python
print('Save...\n')

locations = pd.concat([candidates[country] for country in countries],keys=countries).drop('PLACES',1)
locations.to_pickle(path_to_locations+'locations-'+version_locations+'.pkl')

print('Done!')
```

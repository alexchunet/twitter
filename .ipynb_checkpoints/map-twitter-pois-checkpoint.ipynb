```python
# https://www.kaggle.com/sandeep04201988/eda-of-global-terrorism-using-plotly-and-folium
# https://towardsdatascience.com/data-visualization-with-python-folium-maps-a74231de9ef7
# https://blog.dominodatalab.com/creating-interactive-crime-maps-with-folium/
# https://blogs.oii.ox.ac.uk/bright/2014/08/12/point-size-legends-in-matplotlib-and-basemap-plots/
# https://deparkes.co.uk/2016/06/10/folium-map-tiles/
# https://python-visualization.github.io/folium/quickstart.html
import folium 
import webbrowser
import pandas as pd
import numpy as np
from glob import glob
from timeit import default_timer as timer
from shapely.geometry import Polygon, Point
from folium.plugins import FastMarkerCluster
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
print('Import:')
start = timer()

files = glob('../data/decahose/parsed/locations/places.parquet/*.snappy.parquet')

pois = pd.DataFrame()

for file in files:
    
    pois = pd.concat([pois,pd.read_parquet(file)])

pois = pois[(pois['place_type']=='poi')].copy()
pois.drop(['id','url','place_type','type','name'],1,errors='ignore',inplace=True)
pois.dropna(inplace=True)

pois['coordinates'] = pois['coordinates'].apply(lambda x:Polygon([list(y) for y in x[0]]))
pois['coordinates'] = pois['coordinates'].apply(lambda x:x.centroid)
pois['longitude'] = pois['coordinates'].apply(lambda x:x.x)
pois['latitude'] = pois['coordinates'].apply(lambda x:x.y)

pois = pois.sort_values(by='n_obs',ascending=False).reset_index(drop=True)

end = timer()
print('Computing Time:', round(end - start), 'sec')
```

```python
def barplot_pois():

    with sns.axes_style("white"):
        sns.set_style("ticks")
        sns.set_context("talk")

        fig,ax = plt.subplots(nrows=1,ncols=1,figsize=(30,10))

        to_plot = pois[pois['country_code']!='']['country_code'].value_counts()[:50]
        to_plot.index = to_plot.index.map(lambda x:x.title().replace('_',' '))

        to_plot.plot(
        ax=ax,
        kind='bar',
        color=sns.color_palette("Set2",10))

        ax.tick_params(which='both',direction='in',pad=3)
        ax.locator_params(axis='y',nbins=6)
        ax.set_xticklabels(ax.get_xticklabels(),rotation=45,ha='right')
        ax.set_yscale('log')

    plt.savefig('../fig/barplot-pois.pdf',bbox_inches='tight')
    
barplot_pois()
```

```python
print('Map POIs:')
start = timer()

map_pois = folium.Map(
location = [20, 0], 
zoom_start=2, 
tiles='cartodbpositron',
)
drop=[
'US', 
'JP',
'GB', 
'CA', 
'ES', 
'NL', 
'DE',
'IT', 
'BE', 
'PT',
'FI',
'FR',
'',
]

pois_sample = pois[-pois.country_code.isin(drop)].sample(frac=0.01,weights=pois.n_obs,random_state=1)

for i, (poi, lat, lon, n_users) in enumerate(zip(
pois_sample['full_name'],
pois_sample['latitude'],
pois_sample['longitude'],
pois_sample['n_obs'])):
    
    folium.CircleMarker(
    [lat, lon],
    radius=0.15*np.sqrt(n_users),
    tooltip = poi.title() + ' (' + '{:,}'.format(n_users) + ' Checkins)',   
#     popup = folium.Popup(
#     str(poi).title() + ' (' + '{:,}'.format(n_users) + ' Checkins)', 
#     parse_html=True,
#     max_width=1000), 
    color='b',
    fill=True,
    weight=0,
    fill_opacity=0.5,
    ).add_to(map_pois)
        
map_pois.save('../fig/map-twitter-pois.html')

end = timer()
print('Computing Time:', round(end - start), 'sec')
```

```python
import pandas as pd
from timeit import default_timer as timer
from glob import glob
import pickle
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

# Params

```python
path_to_data = '/scratch/spf248/twitter/data/classification/'
print("path to data:", path_to_data,'\n')

path_to_figs = '/scratch/spf248/twitter/fig/'
print("path to figures:", path_to_figs,'\n')

language = 'en'
print('Language:', language,'\n')

keywords = {
'en':frozenset(['job','position','work','fired','laid off','quit','unemployed','hired']),
}[language]

print("Keywords:", ', '.join(keywords),'\n')

target_sentences = {
0:'I lost my job today',
5:'I was fired earlier this week',
# 'I recently got laid off',
# 'I just quit my job',

1:'Now I am unemployed',
6:'I am currently not working',
    
2:'I am searching for a new position',
7:'Anyone hiring?',
# 'I am looking for a job',
    
3:'I got hired today',
8:'I recently started working at my new job',
# 'I just found a position',
    
4:'Here is a job opportunity you might be interested in',
9:'Looking for a new position?', 
}

labels = [
'1a - DOES THIS TWEET INDICATE THAT THE USER RECENTLY LOST OR LEFT A JOB? [Answer: YES/NO] (Example: I lost my job today — I was fired earlier this week — I recently got laid off — I just quit my job)',
'1b - IF YES: DOES THIS TWEET INDICATE THAT THE USER HAD MULTIPLE JOBS? [Answer: YES/NO] (Example: Can no longer do this multiple jobs thing, I’m quitting — Today I got fired, luckily I still have my other jobs)',
'2 - DOES THIS TWEET INDICATE THAT THE USER IS CURRENTLY UNEMPLOYED? [Answer: YES/NO] (Example: Now I am unemployed — I am currently not working)',
'3a - DOES THIS TWEET INDICATE THAT THE USER IS SEARCHING FOR A JOB? [Answer: YES/NO] (Example: I am looking for a job — I am searching for a new position — Anyone hiring?)',
'3b - IF YES: DOES THIS TWEET INDICATE THAT THE USER IS LOOKING TO HAVE MULTIPLE JOBS? [Answer: YES/NO] (Example: Need to find a second job, anyone hiring? — Not making enough money, need to find a night job)',
'4a - DOES THIS TWEET INDICATE THAT THE USER WAS RECENTLY HIRED? [Answer: YES/NO] (Example: I just found a job — I got hired today — I started working at my new position on Monday — Can’t wait to start my new job!)',
'4b - DOES THIS TWEET INDICATE THAT THE USER HAS MULTIPLE JOBS? [Answer: YES/NO] (Example: Just started my second job today! — Tomorrow I will start my new job. I hope juggling this with my other positions is worth it)',
'5 - DOES THIS TWEET CONTAIN A JOB OFFER? [Answer: YES/NO] (Example: Looking for a new position? — Here is a job opportunity you might be interested in)',
]

print("Labels:\n\n", '\n\n'.join(labels))
```

# Import Data

```python
print('Import Tweets...')
start = timer()

has_keywords = pd.read_pickle(path_to_data+'tweets-keywords-'+language+'.pkl',compression='xz')
is_random    = pd.read_pickle(path_to_data+'tweets-random-'+language+'.pkl',  compression='xz')

print('# Tweets with Keywords:', has_keywords.shape[0])
print('# Random Tweets:', is_random.shape[0])

print("Done in", round(timer()-start), "sec")
```

```python
print("Load Similarity Scores...")
start = timer()

scores = {}

for path in sorted(glob(path_to_data+'tweets-similarity*')):
    
    i_target = int(path.split('-')[-3])
    i_partition = int(path.split('-')[-1].split('.')[0])
    target_sentence = target_sentences[i_target]
#     print('Target Sentence:', target_sentence)
#     print('Partition:', i_partition)
    
    scores[target_sentence + ' - '+ str(i_partition)] = pd.read_pickle(path,compression='xz')

print("Done in", round(timer()-start), "sec")
```

# Merge Data

```python
print("Merge Partition...")
start = timer()

scores = pd.concat([pd.concat([scores[key] for key in scores if target_sentence in key])
for target_sentence in target_sentences.values()],1)

print('# Scores:', scores.shape)

print("Done in", round(timer()-start), "sec")
```

```python
print("Remove Duplicate IDs...")
start = timer()

scores = scores.reset_index().rename(
columns={'index':'ID'}).drop_duplicates('ID').set_index('ID').sort_index()
print('# Scores:', scores.shape[0])

has_keywords = has_keywords.reset_index().rename(
columns={'index':'ID'}).drop_duplicates('ID').set_index('ID').sort_index()
print('# Tweets With Keywords:', has_keywords.shape[0])

is_random = is_random.reset_index().rename(
columns={'index':'ID'}).drop_duplicates('ID').set_index('ID').sort_index()
print('# Random Tweets:', is_random.shape[0])

print("Done in", round(timer()-start), "sec")
```

```python
print("Merge Tweets and Their Scores...")
start = timer()

has_keywords = has_keywords.join(scores,on='ID',how='left')
print('# Tweets With Keywords:', has_keywords.shape)

is_random = is_random.join(scores,on='ID',how='left')
print('# Random Tweets:', is_random.shape)
del scores

print("Done in", round(timer()-start), "sec")
```

# Find Keywords

```python
print("Extract Indices of Tweets Containing Keywords...\n")
start = timer()

idx_keywords = {}
idx_random = {}

for keyword in keywords:
    
    print(keyword)
    
    idx_keywords[keyword] = has_keywords.loc[has_keywords['TEXT'].apply(lambda x:keyword in x)].index
    print('# Matched Tweets:', idx_keywords[keyword].shape[0])
    
    idx_random[keyword] = is_random.loc[is_random['TEXT'].apply(lambda x:keyword in x)].index
    print('# Random Tweets:', idx_random[keyword].shape[0])
    
    print()

print("Done in", round(timer()-start), "sec")
```

# Sample Tweets to be Labeled

```python
print("Send Tweets Containing Keywords to Labeling...\n")
start = timer()

# Keep Track of Tweets Already Sent to Labeling
to_drop = pd.Index([])

to_excel = {}

for keyword in keywords:

    print(keyword)
    
    # Select Tweets With This Keyword
    this_keyword = has_keywords.reindex(idx_keywords[keyword]).copy()

    sheet_name = keyword
    
    # Send a Random Sample of 100 to Be Labeled
    to_excel[sheet_name] = this_keyword.drop(to_drop,errors='ignore')[['TEXT']].sample(n=100,random_state=0)
    
    # Keep Track of Labeled Tweets
    to_drop = to_drop.union(to_excel[sheet_name].index)

    for label in labels:
        to_excel[sheet_name][label] = ''

    for key in sorted(target_sentences):

        sheet_name=keyword+'-target-'+str(key)
        
        # Send a Top 100 to Be Labeled
        to_excel[sheet_name] = this_keyword.drop(to_drop,errors='ignore').sort_values(
        by=target_sentences[key],ascending=False)[['TEXT']].head(100)

        to_drop = to_drop.union(to_excel[sheet_name].index)

        for label in labels:
            to_excel[sheet_name][label] = ''

print("Done in", round(timer()-start), "sec")
```

```python
print("Send Random Tweets to Labeling and Compute Expected Lift Per Keyword...\n")

start = timer()
    
lift = {}

for key in sorted(target_sentences):
    
    target_sentence = target_sentences[key]
    
    print(target_sentence)
    
    is_random.sort_values(by=target_sentence,ascending=False,inplace=True)
    
    sheet_name='random-target-'+str(key)
    
    to_excel[sheet_name] = is_random.drop(to_drop,errors='ignore')[['TEXT']].head(100)
    
    to_drop = to_drop.union(to_excel[sheet_name].index)

    for label in labels:
        to_excel[sheet_name][label] = ''
        
    for keyword in keywords:
        
        name = keyword+'-'+target_sentence
        
        lift[name] = pd.Series([0]*is_random.shape[0], is_random.index, name='keyword')
        
        lift[name].loc[idx_random[keyword]] = 1
        
        lift[name].reset_index(drop=True,inplace=True)
        
        lift[name].index.name = 'rank'
        
print("Done in", round(timer()-start), "sec\n")
```

# Save

```python
print("Store Tweets to be Labeled in Pickle...\n")
start = timer()

with open(path_to_data+'tweets-to-label.pkl','wb') as f:
    
    pickle.dump(to_excel,f)
    pickle.dump(target_sentences,f)

print("Done in", round(timer()-start), "sec\n")
```

```python
print("Store Tweets to be Labeled in Excel...\n")
start = timer()

with pd.ExcelWriter(path_to_data+'tweets-to-label.xlsx') as writer:
    for sheet_name in to_excel:
        to_excel[sheet_name].to_excel(writer,sheet_name=sheet_name)

print("Done in", round(timer()-start), "sec\n")
```

# Analyze Lift Factors

```python
base_rates = pd.Series([is_random.loc[idx_random[keyword]].shape[0]/is_random.shape[0] for keyword in keywords], 
index=keywords).sort_values(ascending=False)
```

```python
base_rates
```

```python
def get_table(target_sentence):
    
    tables = {}
    
    for cutoff in [50000,100000,200000]:
        
        idx = [keyword for keyword in base_rates.index if keyword not in target_sentence.lower()]

        tables[cutoff] = pd.DataFrame([[
        lift[keyword+'-'+target_sentence].head(cutoff).mean(),
        lift[keyword+'-'+target_sentence].head(cutoff).sem()*2]/\
        base_rates[keyword] for keyword in idx],
        index=[keyword+' ('+"{:.1E}".format(base_rates[keyword])+')' for keyword in idx],
        columns=['mean','se'])
        
    tables = pd.concat([tables[key] for key in sorted(tables)],1,keys=sorted(tables))
    
    tables = tables.unstack()
    
    tables.index.names = ['cutoff','stats','keyword']
   
    return tables.unstack(level=['cutoff','stats'])

get_table(target_sentence)
```

```python
width  = 0.2

for key in sorted(target_sentences):

    target_sentence = target_sentences[key]
    
    with sns.axes_style("white"):
        sns.set_style("ticks")
        sns.set_context("talk")

        fig,ax = plt.subplots(nrows=1,ncols=1,figsize=(15,8))

        table = get_table(target_sentence)

        cutoffs = table.columns.get_level_values('cutoff').unique()

        for i,cutoff in enumerate(cutoffs):

            y    = table[cutoff]['mean']
            yerr = table[cutoff]['se']
            x    = np.arange(len(y))

            bars = ax.bar(
            x+i*width, 
            y, 
            width, 
            yerr=yerr,
            color=sns.cubehelix_palette(len(cutoffs))[i],
            error_kw=dict(ecolor='k', lw=1, capsize=1, capthick=1),
            label='{:,}'.format(cutoff)+' tweets')

        ax.legend(loc='best')    
        plt.axhline(y=0, color='grey', linestyle='--', linewidth=1,)
        ax.tick_params(which='both',direction='in',pad=3)
        ax.locator_params(axis='y',nbins=5)
        ax.set_title(target_sentence,fontweight='bold')
        ax.set_xticks(range(len(table.index)))
        ax.set_xticklabels(list(table.index),rotation=30,ha='right')
        ax.axhline(y=1, color='k', linestyle='--', linewidth=1, zorder=0)
        plt.gca().set_ylim(bottom=0)
        ax.set_ylabel('Lift',fontweight='bold')
        
    plt.savefig(path_to_figs+target_sentence.lower().replace(' ','-')+'.pdf',bbox_inches='tight')
```

```python
cutoff = 200000

for key in sorted(target_sentences):
    
    target_sentence = target_sentences[key]
    
    print(target_sentence)

    idx = [keyword for keyword in base_rates.index if keyword not in target_sentence.lower()]

    to_plot = pd.DataFrame([[
    lift[keyword+'-'+target_sentence].head(cutoff).mean(),
    lift[keyword+'-'+target_sentence].head(cutoff).sem()*2]/\
    base_rates[keyword] for keyword in idx],
    index=[keyword+' ('+"{:.1E}".format(base_rates[keyword])+')' for keyword in idx],
    columns=['mean','se'])

    with sns.axes_style("white"):
        sns.set_style("ticks")

        fig,ax = plt.subplots(figsize=(8,5))

        to_plot.plot(
        ax=ax,
        kind='bar',
        y='mean',
        yerr='se',
        legend=None,
        color=sns.color_palette("hls", base_rates.shape[0]),
        error_kw=dict(ecolor='black',elinewidth=0.5,capsize=3,capthick=0.5),
        width=0.5,
        alpha=0.9)

    ax.set_title(target_sentence,fontweight='bold')

    y_label = 'Lift ('+'{:,}'.format(cutoff)+' tweets)'

    ax.set_ylabel(y_label,fontweight='bold')

    ax.set_xlabel('',fontweight='bold')

    ax.tick_params(which='both',direction='in',pad=3)

    ax.locator_params(axis='y',nbins=5)

    ax.set_xticklabels(ax.get_xticklabels(),rotation=45,ha='right')

    ax.axhline(y=1, color='k', linestyle='--', linewidth=1, zorder=0)
    
    plt.gca().set_ylim(bottom=0)

    plt.savefig(path_to_figs+(y_label+'-'+target_sentence).lower().replace(' ','-')+'.pdf',bbox_inches='tight')
```

```python
for key in sorted(target_sentences):
    
    target_sentence = target_sentences[key]
    
    print()
    print(target_sentence)

    with sns.axes_style("white"):
        sns.set_style("ticks")

        fig,ax = plt.subplots(figsize=(8,5))

        for i,keyword in enumerate(base_rates.index):

            if keyword in target_sentence.lower():

                print('Skip', keyword)
                continue

            else:

                name = keyword+'-'+target_sentence

                cutoff = int(min(100/base_rates[keyword], 100000))
            
                bins = np.arange(0,lift[name].shape[0],cutoff)

                to_plot = lift[name].reset_index().groupby(pd.cut(lift[name].index,bins)).agg(['mean','sem'])
                                
                to_plot = (to_plot.set_index(('rank','mean'))['keyword']/base_rates[keyword]).loc[:1000000]
                
                to_plot.plot(
                y='mean',
                yerr='sem',
                ax=ax,
                linewidth=1,
                label=keyword+' (Base Rate:'+"{:.1E}".format(base_rates[keyword])+')',
                color=sns.color_palette("hls", 2*base_rates.shape[0])[2*i])
                
    ax.set_xlim([1, 1000000])
    ax.set_ylim([-20, 70])

    ax.legend(loc='best')

    ax.set_title(target_sentence,fontweight='bold')

    ax.set_ylabel('Lift',fontweight='bold')

    ax.set_xlabel('Rank',fontweight='bold')

    ax.tick_params(which='both',direction='in',pad=3)

    ax.locator_params(axis='both',nbins=5)
    
    plt.savefig(path_to_figs+'lift-'+target_sentence.lower().replace(' ','-')+'.pdf',bbox_inches='tight')
```
